```python
import deepdish as dd
import numpy as np
from brainiak.eventseg import event
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.spatial.distance import jensenshannon
from scipy.stats import zscore
import time
```

```python
# Loading data created by LoadData.ipynb
t0 = time.time()
PMC_h5 = dd.io.load('data/PMC.h5')
PMC = PMC_h5["PMC"]
mask = PMC_h5["mask"]>0
nSubj = PMC.shape[2]
emo = np.load('data/emotions.npy')
emotion_names = np.array(['Anger', 'Contempt', 'Disgust', 'Elation', 'Envy', 'Fear', 'Guilt', 'Hope', 'Interest', 'Joy', 'Pride', 'Relief', 'Sadness', 'Satisfaction', 'Shame', 'Surprise'])
```

# Emotional dynamics

We first look at how the pattern of emotions changes across the movie. At each timepoint, we have a 16-dimensional emotion representation, based on the degree to which each of 16 emotions was present at the timepoint (averaged across observers). We can correlate these patterns for all pairs of timepoints, in order to identify periods during which emotions are relatively stable (indicated by bright blocks along the diagonal).

We also plot how "Interest" varies over the movie, on the same x axis.

```python
plt.figure(figsize=(6,12))
f, (ax1, ax2) = plt.subplots(2,1,figsize=(6,10), gridspec_kw={'height_ratios': [5,1]})

# Plot correlation matrix for (z-scored) emotions
im1 = ax1.imshow(np.corrcoef(zscore(emo.mean(2), axis=1).T))
ax1.set_xlabel('Timepoint')
ax1.set_ylabel('Timepoint')
f.colorbar(im1, ax=ax1, location='bottom', label='Emotion correlation (r)', pad=0.07)

# Plot Interest timecourse
ax2.plot(np.arange(emo.shape[1]), emo[emotion_names == 'Interest',:,:].squeeze().mean(1))
ax2.set_ylabel('Interest')
ax2.set_yticks(ticks=[])
ax2.set_xlabel('Timepoint');
ax2.set_xlim(0, emo.shape[1])

plt.tight_layout()
```

There is visible "event-like" structure along the diagonal, with periods of emotional stability puncutated by rapid transitions to a new emtion. We can use the brainiak Event Segmentation model to identify these emotional shifts. We use a split-half training/testing procedure to identify the number of events for the model that generalize best across the two halves. (Note that the two groups of subjects are not entirely independent here, due to the collaborative filtering approach that mixes data across subjects, so this may slightly overestimate the degree of generalization to independent subjects.)

```python
k_array = np.arange(2,50,5)
test_ll = np.zeros(len(k_array))
nsubj_emo = emo.shape[2]

# Loop over possible settings for the number of events k
print('Trying events:', end=' ')
for i, k in enumerate(k_array):
    print(k, end=' ')
    
    # Fit HMM to half of the subjects
    emo_train = zscore(emo[:,:,:(nsubj_emo//2)].mean(2), axis=1).T
    emo_HMM = event.EventSegment(k)
    emo_HMM.fit(emo_train)
    
    # Test log-likelihood (model fit) on other half of subjects
    emo_test = zscore(emo[:,:,(nsubj_emo//2):].mean(2), axis=1).T
    _, test_ll[i] = emo_HMM.find_events(emo_test)

best_k = k_array[np.argmax(test_ll)]
print('\nOptimal # of events: ', best_k)

# Do a final fit with the optimal number of events, using all subjects
# Set split_merge=True to prioritize fit quality over speed
es = event.EventSegment(best_k, split_merge=True)
es.fit(zscore(emo.mean(2), axis=1).T)
```

The model produces two outputs: a probabilistic assignment of timepoints to events, and the emotion pattern for each event. We can define hard event boundaries by assigning each timepoint to its maximum-probability event, and looking for switches between these events.

```python
# Identify event boundaries as timepoints when the
# maximum-probability event switches
eb = np.where(np.diff(np.argmax(es.segments_[0],axis=1)))[0]

# Plot timepoint-timepoint correlation matrix
_, (ax1, ax2) = plt.subplots(2,1, figsize = (10,20))
ax1.imshow(np.corrcoef(zscore(emo.mean(2), axis=1).T), cmap = 'viridis')
ax1.set_title('Event segmentation for emotions')
ax1.set_xlabel('Timepoint')
ax1.set_ylabel('Timepoint')

# Plot boundaries as boxes on top of correlation matrix
bounds_aug = np.concatenate(([0], eb, [emo.shape[1]]))
for i in range(len(bounds_aug) - 1):
    rect = patches.Rectangle(
        (bounds_aug[i], bounds_aug[i]),
        bounds_aug[i+1] - bounds_aug[i],
        bounds_aug[i+1] - bounds_aug[i],
        linewidth = 2, edgecolor = 'w',facecolor = 'none'
    )
    ax1.add_patch(rect)

# Visualize the emotion patterns for each event
ax2.imshow(es.event_pat_)
ax2.set_yticks(ticks=np.arange(16), labels=emotion_names)
ax2.set_xticks(ticks=np.arange(best_k))
ax2.set_xlabel('Event')

# Print the timing for the events
print('Emotional events:')
for i,b in enumerate(np.concatenate(([0], eb))):
    print('Event', i, ':', time.strftime('%M:%S', time.gmtime((b)*2)))
```

For the rest of the project, we decided to limit our analysis to the final 5 events, since this is where we see the strongest emotional event patterns (and the highest level of interest).

```python
T0 = eb[16]
T = emo.shape[1]-T0
PMC = PMC[:,T0:,:]
emo = zscore(emo[:,T0:,:].mean(2),axis=1)
```

# Measuring neural synchronization across subjects

Do all viewers have synchronized internal representations of the movie? To answer this question, we can fit an event segmentation model across subjects, which assumes:
* All subjects proceed through the same sequence of event states
* Each state corresponds to a specific pattern of voxel activity in PMC
* The timing of these transitions can vary across subjects

We set the number of events so that each event covers approximately 30 seconds of the stimulus. To validate that this fit is meaningful and consistent, we also compute fits on the first and second half of the subjects separately.

```python
# This code block will take at least several minutes to run

# Define events at a 30-second timescale
K = round(T*2/30)

print('Fitting with all subjects...')
es_PMC = event.EventSegment(K)
es_PMC.fit([PMC[:,:,s].T for s in range(nSubj)])

print('Fitting with on first half...')
es_PMC_g1 = event.EventSegment(K)
es_PMC_g1.fit([PMC[:,:,s].T for s in range(nSubj//2)])

print('Fitting with on second half...')
es_PMC_g2 = event.EventSegment(K)
es_PMC_g2.fit([PMC[:,:,s].T for s in range(nSubj//2,nSubj)]);
```

The model identifies a sequence of event patterns shared by all subjects, and the timing of how each subject moves through those patterns.

First, let's look at the event patterns, which are patterns of activation across PMC voxels. We can visualize the correlation matrix between these patterns, and also visualize what the event patterns look like in PMC (for one sagittal slice).

```python
# Plot spatial correlation between HMM-identified events
plt.figure(figsize=(10,10))
plt.imshow(np.corrcoef(es_PMC.event_pat_.T))
plt.xlabel('Event')
plt.ylabel('Event')
plt.colorbar(label='Voxel pattern correlation (r)');
```

```python
# Plot spatial pattern for each event, for one slice through PMC
LRslice = 45
pat2d = np.zeros((mask.shape[0], mask.shape[1], mask.shape[2], K))
pat2d[mask] = es_PMC.event_pat_
pat2d = pat2d[LRslice]
maskvox = np.where(mask[LRslice,:,:])
pat2d = pat2d[np.min(maskvox[0]):np.max(maskvox[0]),np.min(maskvox[1]):np.max(maskvox[1]),:]

f, axes = plt.subplots(5,5, constrained_layout=True, figsize=(10,10))
maxval = 0.8*np.max(np.abs(pat2d))
for i, ax in enumerate(axes.flatten()):
    im = ax.imshow(pat2d[:,:,i].T, interpolation='none',
              origin='lower', cmap='PRGn', vmin = -1*maxval, vmax = maxval)
    ax.axis('off')
    ax.set_title('Event ' + str(i))
f.colorbar(im, ax=axes[4, :], shrink=0.8, location='bottom', label='Voxel activation')
```

The model also outputs a probability distribution over events for every timepoint for every subject. For example, we can plot these probabilities for one subject, showing that at each timepoint the model probabilities are concentrated on 1-3 events at a time.

```python
plt.imshow(es_PMC.segments_[9].T, aspect='auto', interpolation='none', origin='lower')
plt.colorbar(label='p(Event k at timepoint t)')
plt.clim(0,0.6)
plt.xlabel('Timepoint')
plt.xticks(ticks=np.arange(0,T,50),
           labels=[time.strftime('%M:%S', time.gmtime((b)*2)) for b in np.arange(T0,T0+T,50)])
plt.ylabel('Event');
plt.title('One subject')
```

We can compare across subjects to see how these distributions differ. One way of comparing subjects is to measure how far apart they are in time: some subjects are "leading" (transitioning sooner into events) and others are "lagging" (transitioning later into events). We can plot, for each subject, whether they are ahead or behind the average of the group at each timepoint. To check that these estimates are stable across subsets of the dataset, we compute the average lead/lag for each subject using a) the full dataset, and b) half the dataset.

Plotting the fit for the full dataset (each line corresponds to one subject), we see that some subjects vary in their timing on a scale of ~30 seconds, with some consistency in which subjects are ahead of others.

```python
# Compute expected value of the event distribtion at each timepoint
# This indicates how far along the event sequence each subject is
expval = [es_PMC.segments_[s]@np.arange(K) for s in range(nSubj)]
expval_g1 = [es_PMC_g1.segments_[s]@np.arange(K) for s in range(nSubj//2)]
expval_g2 = [es_PMC_g2.segments_[s]@np.arange(K) for s in range(nSubj//2)]
```

```python
# As an example, we can plot one subject's expected value
# vs. the group average
_, ax = plt.subplots(figsize=(6,6))
ax.plot(expval[9])
ax.plot(np.mean(expval,axis=0), 'k--')
ax.legend(['One subject','Average of all subjects'])

ax.set_xticks(np.arange(0,T,50))
ax.set_xticklabels([time.strftime('%M:%S', time.gmtime((b)*2)) for b in np.arange(T0,T0+T,50)])
ax.set_xlabel('Timepoint')
ax.set_ylabel('Expected value of HMM event')
```

```python
# For each subject, compare where they are in the event sequence vs.
# the average across all subjects
lead_lag = np.zeros((nSubj,T))
lead_lag_sh = np.zeros((nSubj,T))

# To get a result in units of seconds: a subject that jumped immediately
# to the last event would be ahead by (K-1) events, which should
# correspond to leading by the entire timecourse = T timepoints * 2 sec/TR
# This gives a conversion factor of (T*2/(K-1))
for s in range(nSubj):
    lead_lag[s,:] = (expval[s] - np.mean(expval,axis=0))*(T*2/(K-1))
    if s < nSubj//2:
        lead_lag_sh[s,:] = (expval_g1[s] - np.mean(expval_g1,axis=0))*(T*2/(K-1))
    else:
        lead_lag_sh[s,:] = (expval_g2[s-nSubj//2] - np.mean(expval_g2,axis=0))*(T*2/(K-1))

_, ax = plt.subplots(figsize=(10,10))
ax.set_prop_cycle(plt.cycler(color=plt.cm.turbo(np.linspace(0, 1, nSubj))))
ax.plot(lead_lag.T);
ax.set_xticks(np.arange(0,T,50))
ax.set_xticklabels([time.strftime('%M:%S', time.gmtime((b)*2)) for b in np.arange(T0,T0+T,50)])
ax.set_xlabel('Timepoint')
ax.set_ylabel('Leading/lagging (seconds)')
ax.set_title('Subject temporal alignment')
ax.set_ylim(-55,55)
ax.plot([0,T],[0,0],'k--')
```

We can obtain a single lead/lag value for each subject by averaging across timepoints, and compare this lead/lag value for each subject when using all subjects vs. using only half of the subjects. We find that these values are similar across the two fits, providing evidence that these estimates are stable measures of individual differences.

```python
f, ax = plt.subplots(figsize=(6,6))
plt.scatter(lead_lag.mean(1),lead_lag_sh.mean(1), c=plt.cm.turbo(np.linspace(0, 1, nSubj)))
plt.plot([-25,25],[-25,25], color='k')
plt.axis('equal')
ax.set_aspect('equal', 'box')
plt.xlim(-25,25)
plt.ylim(-25,25)
plt.xlabel('Lead/lag in full dataset (seconds)')
plt.ylabel('Lead/lag in half dataset (seconds)')
print('Correlation:', np.corrcoef(lead_lag.mean(1),lead_lag_sh.mean(1))[0,1])
```

We can also compute an overall measure of subject synchronization at each timepoint, by measuring the extent to which the event distributions are the same across subjects. We specifically use Jensen-Shannon divergence between each pair of subjects, which ranges from 0 (identical distributions) to 1 (non-overlapping distributions). This measure should increase whenever subjects become more de-synchronized (are in different events at the same timepoint).

```python
def JS(segments_):
    state_dist=np.zeros(T)
    for t in range(T):
        dist_t = []
        for s1 in range(len(segments_)):
            for s2 in range(len(segments_)):
                # For each pair of subjects, measure the
                # divergence between event probabilities
                # at this timepoint 
                dist = jensenshannon(segments_[s1][t,:],
                                     segments_[s2][t,:], base=2)
                dist_t.append(dist)
        avg_dist=np.mean(dist_t)
        state_dist[t] = avg_dist
    return state_dist

js = JS(es_PMC.segments_)
js_g1 = JS(es_PMC_g1.segments_)
js_g2 = JS(es_PMC_g2.segments_)
```

We can plot the divergence for the full fit, as well as the two independent halves of the dataset. We see similar features across all three fits, with a notable peak in divergence between approximately 41:00 and 43:00. This peak indicates a moment in which subjects are most desynchronized, with some subjects substantially ahead of others in terms of their internal event representations.

```python
# JS will be 0 at the beginning/end of the fit due to
# the model assumptions (that all subjects begin/end
# at the same event), so we trim the very first/last
# timepoints which are uninformative
trim = 25

plt.figure(figsize=(8,5))
plt.plot(np.arange(trim, T-trim), js[trim:-trim], color='k', linewidth=2)
plt.plot(np.arange(trim, T-trim), js_g1[trim:-trim])
plt.plot(np.arange(trim, T-trim), js_g2[trim:-trim])
plt.xlabel('Timepoint')
plt.xticks(ticks=np.arange(trim, T-trim,50),
           labels=[time.strftime('%M:%S', time.gmtime((b)*2)) for b in np.arange(T0+trim,T0+T-trim,50)])
plt.ylabel('JS Divergence')
plt.legend(['All subjects', 'First half', 'Second half'])
```

This variability in synchronization correlates closely with "Interest" as reported by the behavioral ratings, with increased interest corresponding to more divergence (i.e. increased interest corresponded to desychronization across subjects). This correlation is present in both subsets of subjects and in the fit to the full dataset.

```python
# Shift JS earlier in time to account for BOLD HRF
HRF_shift = 3

interest_trim = emo[emotion_names == 'Interest',trim:-trim].squeeze()
JS_trim = js[(trim+HRF_shift):(-trim+HRF_shift)]
JS_g1_trim = js_g1[(trim+HRF_shift):(-trim+HRF_shift)]
JS_g2_trim = js_g2[(trim+HRF_shift):(-trim+HRF_shift)]

fig, ax1 = plt.subplots(figsize=(8,5))
ax1.plot(np.arange(trim, T-trim), interest_trim)
ax1.set_ylabel('Interest', color='C0')
ax1.tick_params(axis='y', color='C0', labelcolor='C0')
ax1.set_xlabel('Timepoint')
ax1.set_xticks(np.arange(0, T,50))
ax1.set_xlim(-18.8, 394.8)
ax1.set_xticklabels([time.strftime('%M:%S', time.gmtime((b)*2)) for b in np.arange(T0,T0+T,50)])

ax2 = ax1.twinx()
ax2.plot(np.arange(trim, T-trim), JS_trim, color='k')
ax2.set_ylabel('JS Divergence')

plt.savefig('figs/JS_interest.pdf')

print('Correlation:', round(np.corrcoef(interest_trim, JS_trim)[0,1],2))
print('  Correlation in first half of subjects:', 
      round(np.corrcoef(interest_trim, JS_g1_trim)[0,1],2),
      '  Correlation in second half of subjects:', 
      round(np.corrcoef(interest_trim, JS_g2_trim)[0,1],2))
```
