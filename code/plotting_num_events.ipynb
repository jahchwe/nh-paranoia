```python
mask_dir = '/home/jovyan/paranoia_project/nh-paranoia/code/data/'

label_node = pd.read_csv(mask_dir + '_masks/Schaefer2018_100Parcels_7Networks_order_FSLMNI152_1mm.Centroid_RAS.csv')
label_node['ROI Name']
```

```python
T1 = 526
T2 = 452
T3 = 332
```

```python
# run = 3
# num_events = np.load(save_dir + f'events_per_roi_run{run}_HMM_avg_zscore.npy',allow_pickle=True).item().values()

run1 = np.load('/home/jovyan/paranoia_project/nh-paranoia/schaef_avg_events/event_boundaries_run1_schaeffer_HMM_avg.npy',allow_pickle=True).item()
run2 = np.load('/home/jovyan/paranoia_project/nh-paranoia/schaef_avg_events/event_boundaries_run2_schaeffer_HMM_avg.npy',allow_pickle=True).item()
run3 = np.load('/home/jovyan/paranoia_project/nh-paranoia/schaef_avg_events/event_boundaries_run3_schaeffer_HMM_avg.npy',allow_pickle=True).item()

#need to add the number of TRs to every boundary in runs 2 and 3 


total_bounds = {}
for roi in list(label_node['ROI Name']):
    run2[roi] = [T1 + val for val in run2[roi]]
    run3[roi] = [T1 + T2 + val for val in run3[roi]]
    
    total_bounds[roi] = np.concatenate([run1[roi],run2[roi],run3[roi]])

num_events = [len(val) for val in total_bounds.values()]
```

```python
mask_dir = '/home/jovyan/paranoia_project/nh-paranoia/code/data/'
atlas_fname = (mask_dir + '_masks/'+ 'Schaefer2018_100Parcels_7Networks_order_FSLMNI152_2mm.nii.gz')

mask = Brain_Data(atlas_fname)
mask_x = expand_mask(mask)

mask.plot()
```

```python
brain = roi_to_brain(pd.Series(list(num_events)), mask_x)
brain.plot(cmap='RdBu_r',vmax=np.max(list(num_events)))
```
